The formula that describes the volume V of a hemisphere with radius r is the following:

$ V=\frac{2}{3}\ast\ \pi\ast\ r^3 $

Write a program to calculate the value of V when r is in the range of even numbers from 1-20. Output the result in the following format:

	r	V
	2	…
	4	…
	6	…
	.
	.
	.
	20	…


```python
# Your code here
# importing libraries
import numpy as np

# creating an empty list
r = []

# creating a loop to filter even numbers between 1 and 21 and adding it to the list
for i in np.arange(1,21):
  if i%2 == 0:
    r.append(i)

# printing the headers
print('r\t'+'V')

# loop for printing the each element of r and its corresponding V
for i in r:
  V = (2/3)*np.pi*(i**3)
  print(f"{i}\t{V}")
```

Sentence1: a:2, e:2, i:0, etc.

```python
# import necessary libraries
import re

# convert sentence to list of words
def sentence_to_list(sentence):
  return [char for char in sentence if char.strip()]

# getting valid words from the sentence
def words_list(sentence):
    return re.findall(r"\b\w+\b", sentence)

# getting comman words between the two lists
def common_words(sentence1, sentence2):
  words1 = set(words_list(sentence1))
  words2 = set(words_list(sentence2))
  return list(words1.intersection(words2))

# remove punctuations from the sentences
def remove_punctuation(sentence):
  return re.sub(r"[.,;:!?]", "", sentence)

# count the number of vowels in each sentence
def count_vowels(sentence):
  vowels = "aeiou"
  vowel_count = {v: len(re.findall(v, sentence.lower())) for v in vowels if re.search(v, sentence.lower())}
  return vowel_count

# (a) Get user input
sentence1 = input("Enter the first sentence: ")
sentence2 = input("Enter the second sentence: ")

# (b) Convert sentences to lists of words
sentence_list = sentence_to_list(sentence1) + sentence_to_list(sentence2)
print("(b) Transform each sentence into a list, where each element corresponds to a word in the sentence")
print("List of words:", sentence_list)

# (c) Find common words
common = common_words(sentence1, sentence2)
print("(c) Print out a list of words that occur in both sentences")
print("Common words in both sentences:", common)

# (d) Remove punctuation
sanitized_sentence1 = remove_punctuation(sentence1)
sanitized_sentence2 = remove_punctuation(sentence2)
print("(d) Remove any punctuation from the sentence and print the sanitized sentences")
print("Sanitized Sentence 1:", sanitized_sentence1)
print("Sanitized Sentence 2:", sanitized_sentence2)

# (e) Count vowels
vowels_count1 = count_vowels(sentence1)
vowels_count2 = count_vowels(sentence2)
print("(e) Count the number of vowels in each of the sentences")
print("Vowel occurrences in Sentence 1:", vowels_count1)
print("Vowel occurrences in Sentence 2:", vowels_count2)
```

```python
import math

def evaluate_expression(expression):
  try:
    # Using eval safely by allowing only certain functions and operators
    allowed_names = {"sqrt": math.sqrt, "pow": math.pow, "abs": abs}
    result = eval(expression, {"__builtins__": None}, allowed_names)
    return result
  except ZeroDivisionError:
    return "Error: Division by zero is not allowed."
  except (SyntaxError, NameError):
    return "Error: Invalid input expression."
  except Exception as e:
    return f"Error: {str(e)}"

# main function
def main():
  print("Simple Command-Line Calculator")
  print("Supports +, -, *, /, ** (exponentiation), sqrt(), pow(), and modulus (%)")
  print("Type 'exit' to quit.")

  # while loop to keep on asking an input until exit is specified
  while True:
    expression = input("Enter expression: ").strip()

    if expression.lower() == 'exit':
      print("Exiting calculator. Goodbye!")
      break
      # evaluating the expression and printing the result
    result = evaluate_expression(expression)
    print("Result:", result)

if __name__ == "__main__":
  main()
```

```python
# Filtering data into occupied, vacant and total housing
occupied_df = df[df["title_field"] == "Occupied Housing Units"].sort_values(by="year")
vacant_df = df[df["title_field"] == "Vacant Housing Units"].sort_values(by="year")
total_df = df[df["title_field"] == "Total Housing Units"].sort_values(by="year")
```

```python
# checking if the filtering is done properly
print(occupied_df)
```

```python
'''
(a) Calculate the difference in the number of Occupied Housing Units from year to year and print it.
The difference must be calculated for the consecutive years such as 2008-2009, 2009-2010 etc.
Finally, print the values in the ascending order.
'''

# create new dataframe to store the year and difference
diff_df = pd.DataFrame()
diff_df['year'] = occupied_df['year']
diff_df['value_difference'] = occupied_df['value'].diff()

# resetting te index
diff_df.reset_index(drop=True, inplace=True)

# print the difference values in ascending order
print(diff_df.sort_values(by='value_difference'))
```

```python
'''
(b) For all the years, calculate the percentage of housing units which are vacant an occupied
'''
# copying year into new dataframe
percentage_df = total_df[["year"]].copy()

# calculating the percentages of vacant and occupied units and rounding it to 2 decimals
percentage_df["Vacant Housing Units"] = ((vacant_df["value"].values / total_df["value"].values) * 100).round(2)
percentage_df["Occupied Housing Units"] = ((occupied_df["value"].values / total_df["value"].values) * 100).round(2)

# adding "%" as string, by converting the numerical value to string
percentage_df["Vacant Housing Units"] = percentage_df["Vacant Housing Units"].astype(str) + "%"
percentage_df["Occupied Housing Units"] = percentage_df["Occupied Housing Units"].astype(str) + "%"

# resetting the index numbers
percentage_df.reset_index(drop=True, inplace=True)

# add percentage at the end of values
print(percentage_df)
```

```python
'''
(c) Calculate and print the valued and years in which the highest number of housing units were vacant and occupied.
'''
# getting the max vacant and occupied values
max_vacant = vacant_df.loc[vacant_df["value"].idxmax()]
max_occupied = occupied_df.loc[occupied_df["value"].idxmax()]

# create a dataframe
max_df = pd.DataFrame()

# adding max values to the dataframe
max_df[''] = ['Vacant Housing Units', 'Occupied Housing Units']
max_df['year'] = [max_vacant['year'], max_occupied['year']]
max_df['value'] = [max_vacant['value'], max_occupied['value']]

print(max_df)
```

```python
'''
(d) Calculate the harmonic mean of the total housing units
'''

# Filter the dataframe for Total Housing Units
total_housing = df[df['title_field'] == 'Total Housing Units']

# Calculate the harmonic mean
harmonic_mean = len(total_housing) / np.sum(1 / total_housing['value'])

# printing the harmonic mean
print(f"The harmonic mean of the total housing units is: {harmonic_mean}")
```

#### Edge Case 2
Adding a duplicate ID should not be possible.
#### Prompt
